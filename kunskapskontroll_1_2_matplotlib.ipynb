### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

Plottningen kan antingen göras objektorienterad eller state-baserat/implicit. Det sistnämnda är enklare och kan användas till enklare grafer. Det objektorienterade sättet är rekommenderat för då har man bättre kontroll över alla element i grafen.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure:
En figure är hela ritområdet där visualiseringen skapas. Det kan innehålla en eller flera axes (plottar).

Axes:
Axes är det område inom figuren där data plottas. Det är själva plotten med x- och y-axlar. En figure kan ha flera axes, vilket möjliggör subplots.

Axis:
En axis är en enskild linje som representerar antingen x- eller y-axeln i en axes. Den innehåller skalor, etiketter och titlar för axeln.

Artist: 
En artist är vilket objekt som helst som kan ritas i en figure. Detta inkluderar linjer, punkter, text, axlar, etc. Allt man ser i en Matplotlib-visualisering är en artist.

### When plotting in Matplotlib, what is the expected input data type?

Input data kan vara listor, NumPy-arrays eller Pandas Series.

Listor: plt.plot([1, 2, 3], [4, 5, 6])
NumPy-arrays: plt.plot(np.array([1, 2, 3]), np.array([4, 5, 6]))
Pandas Series: plt.plot(df['x'], df['y'])

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Objektorienterad
fig, ax = plt.subplots()
x = np.linspace(-4, 4, 100)
y = x**2
ax.plot(x, y)
ax.set_title("y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Pyplot
plt.figure()
plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Scatter
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (12,5))
x = np.linspace(-4, 4, 100)
y = x**2

import matplotlib.cm as cm
# Skapa en färgkarta från Matplotlib
colors1 = cm.viridis(np.linspace(0, 1, len(random_data_x)))  # Gradient från colormap

# Scatter plot
ax1.scatter(random_data_x, random_data_y, color=colors1)
ax1.plot(x, y, color='red')

ax1.set_title('Scatter Plot')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# Skapa en färgkarta från Matplotlib
colors2 = cm.viridis(np.linspace(0, 1, len(values)))  # Gradient från colormap

# Bar plot
ax2.bar(names, values, color=colors2)
ax2.set_title('Fruit Quantity')
ax2.set_ylabel('Quantity')

```
